```python
import pandas as pd

df = pd.read_csv('../data/valerius_flaccus/intertext.csv')
df.head()
```

```python
# filter for where col Intertext: Work is Aeneid
df = df[df["Intertext: Work"] == "Aeneid"]
df['VF: Book'] = 1
df.head()
```

```python
df['book'] = 'val_flac'
df['src'] = df['book'] + '.' + df['VF: Book'].astype(str) + '.'+ df['VF: Line Start'].astype(str)

df['tgt_book'] = 'aen_virg'
df['tgt'] =df['tgt_book'] + '.' + df['Intertext: Book'].astype(str) + '.' + df['Intertext: Line Start'].astype(str)
df.head()
```

```python
import glob

vf_files = glob.glob('../data/valerius_flaccus/valerius_flaccus.*', recursive=True)
ae_files = glob.glob('../data/valerius_flaccus/aeneid.*', recursive=True)
```

```python
vf_files
```

```python
vf = []
for f in vf_files[:1]:
    refs = []
    lines = []
    with open(f) as ifd:
        data = ifd.readlines()
        for l in data:
            try:
                ref, line = l.split('\t')
                refs.append(ref)
                lines.append(line.strip().rstrip())
            except:

                ref, line = l.split("   ")
                refs.append(ref)
                lines.append(line.strip().rstrip())

vf = pd.DataFrame({'ref': refs, 'line': lines})
vf.head()
```

```python
vf.iloc[579:589]
```

```python
ae = []
for f in ae_files:
    ae.append(pd.read_csv(f, sep="\t", names=["ref", "line"]))
ae = pd.concat(ae)
ae.head()
```

```python
import re
pattern = r'(\d+)\.(\d+)'


for d in [vf, ae]:
    d['loc'] = d['ref'].apply(lambda x: re.findall(pattern, x)[0])
    d['book'] = d['loc'].apply(lambda x: int(x[0]))

    d['verse'] = d['loc'].apply(lambda x: int(x[1]))

vf.head()
```

```python
# amazing, Latin is supported by LabSE, so we just need to encode it and modify the script to test the IT metric, chatgpt might be able to make it easy work
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/LaBSE')
```

```python
# encode and save vf

# actually, since we assume it's really intratextuality, we should combine these into on df with the location being book:chapter:verse
# and then everything is the same! Just don't want to use the actual Location object
vf_mod = vf[vf['book'] == 1]
```

```python
def mod_cols(df, name):
    df['chapter'] = df['book']
    df['book'] = name
    return df


vf_mod = mod_cols(vf, 'val_flac')
ae_mod = mod_cols(ae, 'aen')
vf_mod.head()
```

```python
ae_mod.head()
```

```python
combined = pd.concat([vf_mod, ae_mod])
combined.head()
```

```python
combined['location'] = combined['book'] + '.' + combined['chapter'].astype(str) + '.' + combined['verse'].astype(str)
```

```python
# first we need to make golds
golds = {k:v for k,v in zip(df['src'], df['tgt'])}
len(golds)  

import pickle
with open("../data/valerius_flaccus/golds.pkl", "wb") as f:
    pickle.dump(golds, f)
```
