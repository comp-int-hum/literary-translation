```python
import pandas as pd

df = pd.read_csv('../data/valerius_flaccus/intertext.csv')
df.head()
```

```python
# filter for where col Intertext: Work is Aeneid
df = df[df["Intertext: Work"] == "Aeneid"]
df['VF: Book'] = 1
df.head()
```

```python
df['book'] = 'val_flac'
df['src'] = df['book'] + '.' + df['VF: Book'].astype(str) + '.'+ df['VF: Line Start'].astype(str)

df['tgt_book'] = 'aen_virg'
df['tgt'] =df['tgt_book'] + '.' + df['Intertext: Book'].astype(str) + '.' + df['Intertext: Line Start'].astype(str)
df.head()
```

```python
import glob

vf_files = glob.glob('../data/valerius_flaccus/valerius_flaccus.*', recursive=True)
ae_files = glob.glob('../data/valerius_flaccus/aeneid.*', recursive=True)
```

```python
vf_files
```

```python
vf = []
for f in vf_files[:1]:
    refs = []
    lines = []
    with open(f) as ifd:
        data = ifd.readlines()
        for l in data:
            try:
                ref, line = l.split('\t')
                refs.append(ref)
                lines.append(line.strip().rstrip())
            except:
                ref, line = l.split("   ")
                refs.append(ref)
                lines.append(line.strip().rstrip())

vf = pd.DataFrame({'ref': refs, 'line': lines})
vf.head()
```

```python
import xml.etree.ElementTree as ET
import csv

# Function to parse the TEI/XML file and extract data
def parse_tei_to_csv(tei_file, output_csv):
    NAMESPACE = {'tei': 'http://www.tei-c.org/ns/1.0'}
    # Parse the XML file
    tree = ET.parse(tei_file)
    root = tree.getroot()

    # Open CSV file for writing
    with open(output_csv, mode='w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        # Write CSV header
        writer.writerow(['reference', 'text'])
        
        # Navigate through the structure of the XML file
        for div in root.findall('.//tei:div[@type="textpart"]', NAMESPACE)[:10]:
            #print(div)
            poetry_part = div.attrib['n']  # Get the "n" attribute (poem number)
            #print(poetry_part)
            # For each line within the poem, extract the line number and text
            for line in div.findall('.//tei:l', NAMESPACE):
                line_number = line.attrib['n']  # Get the "n" attribute (line number)
                print(f"{poetry_part}.{line_number}")
                text = line.text.strip() if line.text else ''  # Get the text inside the <l> tag
                print(text)
                # Create the identifier as "poetry_part.line_number"
                identifier = f"{poetry_part}.{line_number}"
            
                # Write the data to the CSV
                writer.writerow([identifier, text])

# Usage
tei_file = '../data/valerius_flaccus/amores.xml'
output_csv = '../data/valerius_flaccus/amores.csv'

# Call the function to perform the conversion
parse_tei_to_csv(tei_file, output_csv)
```
