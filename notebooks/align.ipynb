```python
import pickle
import sys
import os
import pandas as pd

# want to add ../data to the path so we can import the data
# insert ../data
sys.path.insert(0, '../data')
sys.path.insert(0, "../scripts")
from utils import Location
```

```python
old_BC = pickle.load(open('../data/old_BC_data_with_scores_new_alg.pkl', 'rb'))
new_BC = pickle.load(open('../data/new_BC_data_with_scores_new_alg.pkl', 'rb'))
old_SG = pickle.load(open('../data/old_Sinaiticus-Greek_data_with_scores_new_alg.pkl', 'rb'))
new_SG = pickle.load(open('../data/new_Sinaiticus-Greek_data_with_scores_new_alg.pkl', 'rb'))
old_WLC = pickle.load(open('../data/old_WLC-Hebrew_data_with_scores_new_alg.pkl', 'rb'))
```

```python
from tqdm import tqdm

dfs = []
for k,v in old_SG.items():
    if 'Japanese' not in k:

        # actually want to make a dataframe of location to length
        df = pd.DataFrame.from_records(v)
        # get the number of duplicates
        # print()
        x = len(df)
        df = df.drop_duplicates(subset=['locations'])
        print(f"Dropping {x-len(df)} duplicates for {os.path.basename(k)}")
        # print(len(df)-len(df.drop_duplicates()))
        # let's see the duplicated values
        # print(len(df[df.duplicated()]))
        # print(sum(df['locations'].duplicated() & df['lengths'].duplicated()))
        # print(sum(df['duplicates']))
        # rename the column from lengths to the filename
        df = df.rename(columns={'lengths': os.path.basename(k)})
        dfs.append(df)
        # print(df.head())

# now we want to join the dfs on location
old_SG_df = dfs[0]
for i in tqdm(range(1, len(dfs))):
    old_SG_df = pd.merge(old_SG_df, dfs[i], on='locations', how='outer')


# now we do the same thing with the BC data
dfs_bc = []
for k,v in old_BC.items():
    if 'Japanese' not in k:
        # actually want to make a dataframe of location to length
        df = pd.DataFrame.from_records(v)
        # get the number of duplicates
        # print()
        x = len(df)
        df = df.drop_duplicates(subset=['locations'])
        print(f"Dropping {x-len(df)} duplicates for {os.path.basename(k)}")
        df = df.rename(columns={'lengths': os.path.basename(k)})
        dfs_bc.append(df)

old_BC_SG_df = old_SG_df
for i in tqdm(range(len(dfs_bc))):
    old_BC_SG_df = pd.merge(old_BC_SG_df, dfs_bc[i], on='locations', how='outer')

# change the order so locations (currently column 1) is the first column
old_BC_SG_df = old_BC_SG_df[['locations'] + [col for col in old_BC_SG_df.columns if col != 'locations']]
old_BC_SG_df.head()
```

```python
# same process for old WLC
dfs_wlc = []
for k,v in old_WLC.items():
    if 'Japanese' not in k"
        # actually want to make a dataframe of location to length
        df = pd.DataFrame.from_records(v)
        # get the number of duplicates
        # print()
        x = len(df)
        df = df.drop_duplicates(subset=['locations'])
        print(f"Dropping {x-len(df)} duplicates for {os.path.basename(k)}")
        df = df.rename(columns={'lengths': os.path.basename(k)})
        dfs_wlc.append(df)

old_wlc_df = dfs_wlc[0]
for i in tqdm(range(1, len(dfs_wlc))):
    old_wlc_df = pd.merge(old_wlc_df, dfs_wlc[i], on='locations', how='outer')

# want to merge in dfs_bc
old_BC_WLC_df = old_wlc_df
for i in tqdm(range(len(dfs_bc))):
    old_BC_WLC_df = pd.merge(old_BC_WLC_df, dfs_bc[i], on='locations', how='outer')

# change the order so locations (currently column 1) is the first column
old_BC_WLC_df = old_BC_WLC_df[['locations'] + [col for col in old_BC_WLC_df.columns if col != 'locations']]
print(old_BC_WLC_df.columns)
old_BC_WLC_df.head()
```

```python
# and the same thing for new SG and new BC
dfs_new_sg = []
for k,v in new_SG.items():
    # actually want to make a dataframe of location to length
    df = pd.DataFrame.from_records(v)
    # get the number of duplicates
    # print()
    x = len(df)
    df = df.drop_duplicates(subset=['locations'])
    print(f"Dropping {x-len(df)} duplicates for {os.path.basename(k)}")
    df = df.rename(columns={'lengths': os.path.basename(k)})
    dfs_new_sg.append(df)

new_SG_df = dfs_new_sg[0]
for i in tqdm(range(1, len(dfs_new_sg))):
    new_SG_df = pd.merge(new_SG_df, dfs_new_sg[i], on='locations', how='outer')

# same process for new BC
dfs_new_bc = []
for k,v in new_BC.items():
    # actually want to make a dataframe of location to length
    df = pd.DataFrame.from_records(v)
    # get the number of duplicates
    # print()
    x = len(df)
    df = df.drop_duplicates(subset=['locations'])
    print(f"Dropping {x-len(df)} duplicates for {os.path.basename(k)}")
    df = df.rename(columns={'lengths': os.path.basename(k)})
    dfs_new_bc.append(df)

new_BC_SG_df = new_SG_df
for i in tqdm(range(1, len(dfs_new_bc))):
    new_BC_SG_df = pd.merge(new_BC_SG_df, dfs_new_bc[i], on='locations', how='outer')

# change the order so locations (currently column 1) is the first column
new_BC_SG_df = new_BC_SG_df[['locations'] + [col for col in new_BC_SG_df.columns if col != 'locations']]
new_BC_SG_df.head()
```
