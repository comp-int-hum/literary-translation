```python
import pandas as pd
import random

def combine_random_subset(file1, file2, output_file, books, subset_size):
    # Load the two Excel files into DataFrames
    df1 = pd.read_excel(file1)
    df2 = pd.read_excel(file2)

    # Concatenate the two DataFrames
    combined_df = pd.concat([df1, df2])

    # Filter the DataFrame for rows where 'refs' contain the specified books
    filtered_df = combined_df[combined_df['refs'].str.contains('|'.join(books))]

    # Take a random subset of rows
    if len(filtered_df) > subset_size:
        random_subset = filtered_df.sample(n=subset_size, random_state=1)
    else:
        random_subset = filtered_df  # If there are fewer rows than the subset size

    # Write the random subset to a new Excel file
    
    random_subset.to_excel(output_file, index=False)
    print(f"Random subset saved to '{output_file}'.")
    return random_subset


```

```python
# Example usage
books_to_include = ['Job']
df = combine_random_subset('verse_ngram_chiasm.xlsx', 'verse_neural_chiasm.xlsx', 'job_subset.xlsx', books_to_include, subset_size=20)
df
```
